```python
from pandas_datareader import DataReader
import datetime
import pandas as pd
```

```python
start_date = datetime.date(2015, 1, 1)
end_date = datetime.date(2016, 12, 31)
now = datetime.datetime.now()
ticker = 'GOOG' 
data_source = "yahoo" 
```

```python
stock_data = DataReader(ticker, data_source, start_date, now)
```
